```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score

# 2. Load Dataset
df = pd.read_csv("/content/Fragrance Dataset - COM7022 - [4037] (1).csv")
print("Initial shape:", df.shape)
print(df.head())

# 3. Data Cleaning & Preprocessing
# Drop irrelevant textual columns
df = df.drop(['title', 'priceWithCurrency', 'availableText'], axis=1, errors='ignore')

# --- Convert price ---
def parse_price(x):
    if pd.isna(x): return np.nan
    s = str(x).replace('','').replace('$','').replace(',','').strip()
    try:
        return float(s)
    except:
        return np.nan
df['price'] = df['price'].apply(parse_price)

# --- Convert sold ---
df['sold'] = pd.to_numeric(df['sold'], errors='coerce')

# --- Convert availability to binary ---
df['available'] = df['available'].astype(str).str.lower().map({
    'true':1, 'yes':1, '1':1, 'false':0, 'no':0, '0':0
}).fillna(0)

# --- Convert lastUpdated to "days since update" ---
df['lastUpdated'] = pd.to_datetime(df['lastUpdated'], errors='coerce')
df['days_since_update'] = (pd.Timestamp.now() - df['lastUpdated']).dt.days
df.drop(columns='lastUpdated', inplace=True)

# --- Handle categorical values ---
for col in ['brand','type','itemLocation']:
    df[col] = df[col].fillna('Unknown')

# --- Fill numeric missing values ---
for col in ['price','sold','days_since_update']:
    df[col].fillna(df[col].median(), inplace=True)

# --- Label Encode categorical columns ---
le = LabelEncoder()
for col in ['brand','type','itemLocation']:
    df[col] = le.fit_transform(df[col])

print("\nCleaned dataset sample:\n", df.head())

# 4. EXPLORATORY DATA ANALYSIS (EDA)

# --- Chart 1: Price Distribution ---
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')
plt.title("Price Distribution of Perfumes")
plt.xlabel("Price ()")
plt.ylabel("Frequency")
plt.show()

# --- Chart 2: Scatter Plot (Price vs Sold) ---
plt.figure(figsize=(8,5))
sns.scatterplot(x='price', y='sold', hue='available', data=df, palette='coolwarm', s=50)
plt.title("Price vs Sold Quantity")
plt.xlabel("Price ()")
plt.ylabel("Units Sold")
plt.show()

# --- Chart 3: Correlation Heatmap ---
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='viridis')
plt.title("Correlation Between Numeric Features")
plt.show()

# 5. FEATURE SCALING
features = ['brand','type','itemLocation','price','sold','available','days_since_update']
X = df[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 6. APPLY CLUSTERING MODELS

# --- (a) K-MEANS CLUSTERING ---
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
kmeans_sil = silhouette_score(X_scaled, kmeans_labels)
kmeans_db = davies_bouldin_score(X_scaled, kmeans_labels)

# --- (b) HIERARCHICAL CLUSTERING ---
hc = AgglomerativeClustering(n_clusters=4)
hc_labels = hc.fit_predict(X_scaled)
hc_sil = silhouette_score(X_scaled, hc_labels)
hc_db = davies_bouldin_score(X_scaled, hc_labels)

# --- (c) DBSCAN CLUSTERING ---
dbscan = DBSCAN(eps=0.8, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
mask = dbscan_labels != -1
if len(set(dbscan_labels[mask])) > 1:
    dbscan_sil = silhouette_score(X_scaled[mask], dbscan_labels[mask])
    dbscan_db = davies_bouldin_score(X_scaled[mask], dbscan_labels[mask])
else:
    dbscan_sil, dbscan_db = None, None

# 7. EVALUATION RESULTS
print("\n=== INTERNAL VALIDATION RESULTS ===")
print(f"K-Means:        Silhouette = {kmeans_sil:.3f},  DaviesBouldin = {kmeans_db:.3f}")
print(f"Hierarchical:   Silhouette = {hc_sil:.3f},  DaviesBouldin = {hc_db:.3f}")
if dbscan_sil is not None:
    print(f"DBSCAN:         Silhouette = {dbscan_sil:.3f},  DaviesBouldin = {dbscan_db:.3f}")
else:
    print("DBSCAN:         Could not compute (too few clusters or excessive noise)")

# 8. DIMENSIONALITY REDUCTION (PCA) FOR VISUALIZATION
pca = PCA(n_components=2)
pca_data = pca.fit_transform(X_scaled)

# --- K-Means PCA Visualization ---
plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], hue=kmeans_labels, palette='tab10', s=50)
plt.title("K-Means Clustering Visualization (PCA 2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

# --- Hierarchical PCA Visualization ---
plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], hue=hc_labels, palette='Set2', s=50)
plt.title("Hierarchical Clustering Visualization (PCA 2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

# --- DBSCAN PCA Visualization ---
plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], hue=dbscan_labels, palette='Spectral', s=50)
plt.title("DBSCAN Clustering Visualization (PCA 2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

# 9. CLUSTER PROFILING (BUSINESS INTERPRETATION)
df['KMeans_cluster'] = kmeans_labels
df['Hierarchical_cluster'] = hc_labels
df['DBSCAN_cluster'] = dbscan_labels

# Cluster summary
kmeans_summary = df.groupby('KMeans_cluster')[['price','sold','available','days_since_update']].mean()
print("\n=== K-MEANS CLUSTER PROFILE (AVERAGE VALUES) ===")
print(kmeans_summary)
```
